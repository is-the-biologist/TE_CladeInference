# README:

This notebook will guide the user through calling the functions from haploTE.py to process the copy number and SNP pileup numpy files|

This notebook is the culmination of the full data processing pipeline. We begin with CN and SNP numpy files for each TE. We then use these to create an allele CN for each of the alleles at each position for each strain. Using this CN info we then create a numpy file that contains the sequence diversity estimates for each position in dataset wide and population specific manner. 

### Notes on haploTE module:

There are many additional functions within this module that I have used to generate various plots and analyze data. These may be ignored for the purposes of your own analysis, but for transparency are included in the repo.


```python
from HaploTE import summary_stats, subfamilyInference
```

```python
stats = summary_stats()
hapInference = subfamilyInference()
```

```python
for TE in hapInference.active_fullLength:
    
    stats.calcPi(TE=TE)
```

```python
#Run method:
for TE in hapInference.active_fullLength:
    hapInference.CNAP_extraction(TE)
```
