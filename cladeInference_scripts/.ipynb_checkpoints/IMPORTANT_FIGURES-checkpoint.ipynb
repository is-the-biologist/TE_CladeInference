```python
KW_table = pd.read_csv('/Users/iskander/Documents/Barbash_lab/TE_diversity_data/HAPLOTYPE_CLUSTERS/HAPLOTYPE_CALL_07-17-2020/STATS/KW.RUN_07-11-2020.csv')


max_pop = np.asarray([])
min_pop = np.asarray([])
diff_pop = np.asarray([])
gdl = np.asarray([])
natural_diff_arr = np.asarray([])
for TE in hapAnalysis.active_fullLength[0]:
    hapTable = pd.read_csv(os.path.join(hapAnalysis.haplo_path,f"{TE}.haplotypeTable.tsv"), sep='\t')
    GDL_mean = hapTable["GDL MEAN CN"].values
    population_means = hapTable[["B MEAN CN", "I MEAN CN", "N MEAN CN","T MEAN CN", "Z MEAN CN"]].values
    gdl = np.concatenate((gdl, GDL_mean))
    GDL_mean = GDL_mean.reshape(GDL_mean.shape[0], -1)
     
    abs_diff = abs(GDL_mean - population_means)
    max_diff = np.argmax(abs_diff, axis=1)
    diff_arr = [population_means[j, max_diff[j]] for j in range(population_means.shape[0])]
    
    natural_diffs = np.asarray([ (population_means - GDL_mean) [j][max_diff[j]] for j in range(population_means.shape[0])])
    natural_diff_arr = np.concatenate((natural_diff_arr, natural_diffs))
    
    diff_pop = np.concatenate((diff_pop, diff_arr))
    
    max_pop = np.concatenate((max_pop, np.max(population_means, axis=1)))
    min_pop = np.concatenate((min_pop, np.min(population_means, axis=1)))

KW_table["max_pop"] = max_pop
KW_table["min_pop"] = min_pop
KW_table["GDL_mean"] = gdl
KW_table["diff"] = diff_pop
KW_table["nat_diff"] = natural_diff_arr
bonferonni = -np.log2(0.05 / KW_table.shape[0])
print(bonferonni)
sigs = np.asarray(["n.s." for i in range(KW_table.shape[0])])
sigs[KW_table["P"] > bonferonni] = "sig."
KW_table["sig"] = sigs
```

```python
#adding pseudocounts to reads and CN to be able to log transform values and placate Mike god help me
filtered_df.loc[:,"+"] = filtered_df["+"] + 1
filtered_df.loc[:,"-"] = filtered_df["-"] + 1
filtered_df.loc[:,"CN"] = filtered_df["CN"] + 1
```

```python


clusterNames = []
TE_names = []
prop_indivs = []
num_indivs = []
antisense_reads = []
sense_reads = []
disp_test = []
CN_arr = []
IoD = []
dispersion = []
mean_sense_reads = []
mean_antisense_reads = []
clean_class = {"HTT":"Telomeric", "LINE":"LINE-like", "LTR":"LTR retrotransposon", "TIR":"DNA transposon"}


for group in filtered_df.groupby("TE"):
    poiss_df = hapAnalysis.plot_haplotypeDispersion(group[0])
    for subgroup in group[1].groupby("Cluster"):
        disp = poiss_df[poiss_df["Cluster_ID"] == subgroup[0]]["Poisson fit"].values[0]
        dispersion.append(disp)
        
        if disp == "Dispersed" or disp == "Overdispersed":
            disp_test.append("Young")
        else:
            disp_test.append("Old")
        clust = poiss_df["Cluster_ID"] == subgroup[0]
        index = (poiss_df[clust]["MEAN"]) - (poiss_df[clust]["VAR"])
        
        IoD.append(10**index.values[0])
        TE_names.append(group[0])
        clusterNames.append(subgroup[0])


        mean_anti = np.mean( a=(subgroup[1]["-"]/subgroup[1]["CN"]).values )
        mean_sense = np.mean( a=(subgroup[1]["+"]/subgroup[1]["CN"]).values )
        mean_antisense_reads.append(np.mean( a=(subgroup[1]["-"])))
        mean_sense_reads.append(np.mean( a=(subgroup[1]["+"])))
        
        num_indivs.append(subgroup[1].shape[0])
        antisense_reads.append(mean_anti)
        sense_reads.append(mean_sense)
        mean_CN = np.mean(subgroup[1]["CN"])
        CN_arr.append(mean_CN)
        
        
prop_DF = pd.DataFrame(data={"TE":TE_names+TE_names, "Cluster":clusterNames+clusterNames, "reads/copy":np.asarray(antisense_reads+sense_reads), 
                             "strand":["-" for i in range(len(antisense_reads))] + ["+" for i in range(len(antisense_reads))],
                             "Poiss":disp_test+disp_test, "reads":mean_antisense_reads+mean_sense_reads,
                            "CN":np.asarray(CN_arr+CN_arr), "IoD":IoD+IoD, "dispersion":dispersion+dispersion})

cleanNames = [hapAnalysis.TE_nameTable[N] for N in prop_DF["TE"]]
te_table = pd.read_csv("/Users/iskander/Documents/Barbash_lab/TE_diversity_data/ACTIVE_TES_full.tsv", header=None)
class_table = {te_table[0][j]:te_table[1][j] for j in range(te_table.shape[0])}
classes = [clean_class[class_table[s]] for s in prop_DF["TE"]]
prop_DF["class"] = classes
prop_DF["TE"] = cleanNames
#prop_DF["reads/copy+1"] = prop_DF["reads/copy"] + 1
```

```python
corr_coef = np.asarray([])
p_vals = np.asarray([])
TE_names = np.asarray([])
haplos = np.asarray([])
activity = np.asarray([])
strand = np.asarray([])
copy_number = np.asarray([])
IoD = np.asarray([])
reads = np.asarray([])
poiss_arr = np.asarray([])

for TE in piRNAs.active_fullLength[0]:
    poiss_df = hapAnalysis.plot_haplotypeDispersion(TE)
    putative_activity = np.asarray(["Young" if poiss == "Overdispersed" or poiss == "Dispersed" else "Old" for poiss in poiss_df["Poisson fit"] ])
    iod = (10**poiss_df["MEAN"]) / (10**poiss_df["VAR"])
    
    
    #add for antisense:
    R,P,H = piRNAs.haplotype_piRNA(f"{TE}", stranded="antisense")
    corr_coef = np.concatenate((corr_coef, R))
    p_vals = np.concatenate((p_vals, P))
    haplos = np.concatenate((haplos, H))
    activity = np.concatenate((activity, putative_activity))
    TE_names = np.concatenate( (TE_names, np.asarray([TE for i in range(len(R))])) )
    strand = np.concatenate( (strand, np.asarray(["-" for i in range(len(R))])) )
    copy_number = np.concatenate((copy_number, poiss_df["MEAN"]))
    poiss_arr = np.concatenate((poiss_arr, poiss_df["Poisson fit"].values))
    
    #add for sense:
    R,P,H = piRNAs.haplotype_piRNA(f"{TE}", stranded="sense")
    corr_coef = np.concatenate((corr_coef, R))
    p_vals = np.concatenate((p_vals, P))
    haplos = np.concatenate((haplos, H))
    activity = np.concatenate((activity, putative_activity))
    TE_names = np.concatenate( (TE_names, np.asarray([TE for i in range(len(R))])) )
    strand = np.concatenate( (strand, np.asarray(["+" for i in range(len(R))])) )
    copy_number = np.concatenate((copy_number, poiss_df["MEAN"]))
    poiss_arr = np.concatenate((poiss_arr, poiss_df["Poisson fit"].values))
    
    
piRNA_corr_df = pd.DataFrame(data={"r":corr_coef,"p":p_vals,"TE":TE_names, "dispersion":poiss_arr,
                                   "haplotype":haplos, "poiss":activity, "strand":strand, "CN":copy_number})
```

Alternative to SD, mean of CN distributions could be to calculate skewness. This might not quite work since the distr are probably not normal. 

```python
M = np.full(fill_value=np.nan, shape=(14, 41), dtype="object")
i = 0
for g in haploStats.groupby("TE"):
    CN_norm = g[1]["rawCN"].values
    
    #test for non-param skewness Zheng T, Gastwirth J (2010) "On bootstrap tests of symmetry about an unknown median". Journal of Data Science, 8(3): 413–427
    D = ((2*len(CN_norm))**0.5) * ((np.mean(CN_norm) - np.median(CN_norm)) / (np.var(CN_norm)**0.5))
    p_val = sp.stats.t.sf(D, df=len(CN_norm)-1)
    

    M[:,i] = [g[0], g[1]["TE_class"].values[0], np.mean(CN_norm), np.var(CN_norm)**0.5, sp.stats.iqr(CN_norm),
              np.mean(CN_norm)/(np.var(CN_norm)**0.5), g[1].shape[0], np.percentile(CN_norm, 25), np.percentile(CN_norm, 75), 
              sp.stats.skew(CN_norm, bias=False), sp.stats.kurtosis(CN_norm, fisher=True), (np.mean(CN_norm) - np.median(CN_norm)) / (np.var(CN_norm)**0.5), 
              D, p_val]
    
    i += 1
    
CN_df = pd.DataFrame(data=M.T, columns=["TE", "class", "mean", 'SD', 'iqr', "cv", "N", "q1", "q3", "skew", "kurt", "np_skew", "D", "p"])
```
